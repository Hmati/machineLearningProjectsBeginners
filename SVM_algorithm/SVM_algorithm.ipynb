# INTRODUCTION TO SUPPORT VECTOR MACHINE

Support vector machine learning is a subset of supervised learning algorithm used for classification primarily. It offers high accuracy compared to other algorithms like logistica regression. SVM is used in a variaty of applications such as intrusion detection, email classification, handwritin recognition, face detection and web pages classification

SVM is primarily used for classification problems but can aslo be used for regression problems. It can eaily handle continuous and categorical.

## Terminologies

1. **hyper plane**
    
    Thi is a decision plane which separate a set of objects having different classes. Its basically a boundary that seperates data points, dividing them into classes. Its the half of the value of the margin.
    
    - in one dimension, a hyperplane is called a point
    - in two dimensions, it is a line
    + in three dimensions, it is a plane
    * in more dimensions you can call it an hyperplane
    
    
2. **Support vector**

    These are the data points closest to the hyperplane
    
    
3. **Margin or distance margin**

     This is the distance between the two lines on the support vectors, its the perpendicular line between the support vectors of each class. The bigger the margin the better,we always want to find the highest margin distance.

## How an SVM works

As mentioned earlier the SVM is used for classification, its main aim is to divide the data set into groups(segregate the data). SVM searchs for the maximum marginal hyper plane, let me give you a quick summary of how the SVM  works.

1. Find the hyperplane with the maximum margin


2. incase a nonlinear hyperplane can't be applied due to the datasets being too mixed up, where one class data points are found in the data points of another class. In such cases we use the kernel functions, a kernel function basically takes a low dimension input and outputs a higher dimension dataset. Example it takes a 2-D and outputs a 3-D. This increase in dimention makes it easy to segregate or divide the different classes. This technique is called `kernel trick`. In simply terms the SVM kernel converts nonseparable problem into a separable one by adding more dimensions to it so a plane can be drawn for classification.

## Goal of SVM

The goal of a support vector machine is to find  the optimal separating hyperplane which maximizes the margin of the training data.

# Parameter Tunning

Choosing the right kernelto use is very crucial a bad decision will lead to  a poorly performing model. Always check if your data is linear then use a **linear kernel**(linear SVM). Linear SVM is a parametric model, but an RBF kernel SVM isnâ€™t, so the complexity grows with the size of the training set. The RBF kernel SVM are more expensive and you have to also store the kernel matrix in memory, as the dimensions increase, it becomes computationaly expensive and overfitting the model is an easy occurance.


### 1.  Regularization

This parameter tells the SVM algorithm how much miss classification you want to avoid. This parameter is represented by the capital letter **C**.

The higher **C** the model will use a smaller hyperplane margin, so the model can increase avoidance of miss classification of data points

The lowe the **C** the model will use a maximum margin hyperplane, so as to increase the tolerance of miss classification of data points.

In short the **reguralization value is directly proportional to the margin size**.

![](images/image6.png)

Using the best regularization value is key to avoid overfitting the model.

### 2.  Gamma

Gamma is another important parameter, it determines how far away and how many support vectors are used to calculate the hyperplane. Higher gamma means the model will use support vectors that are closest to the plausible hyperplane while low gamma means support vectors at great distances will be used.  

![](images/image7.png)

In simple terms, gamma is inversely proportional to the number and distance of the support vectors used to determine the plausible hyperplane.


### 3. Margin

The margin is another important parameter to take into account, when using SVM the margin we want is the one with the highest value(maximum margin hyperplane). The higher the margin the better the model or classification.

Here, maximizing the distances between nearest data point and hyper-plane will help us to decide the right hyper-plane to use.

## Let's do some real machine learning

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer
```

```python
breast_cancer = load_breast_cancer()
```

```python
breast_cancer
```

```python
df = pd.DataFrame(breast_cancer.data, columns=breast_cancer.feature_names)
df
```

```python
df['target'] = breast_cancer.target
df
```

```python
df['target_name'] = pd.Categorical.from_codes(breast_cancer.target, breast_cancer.target_names)
df
```

## Splitting the data

Now that we have done some data exploration, lets dive into splitting the data set into traininng and testing sets

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(breast_cancer.data, breast_cancer.target, train_size=0.8)
```

```python
from sklearn.svm import SVC
```

```python
sup_vec = SVC(kernel='linear')
```

```python
sup_vec.fit(X_train, y_train)
```

```python
predictions = sup_vec.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test, predictions)
```

```python
from matplotlib import pyplot as plt
import seaborn as sn
```

```python
plt.figure(figsize=(12, 7))
sn.heatmap(cm, annot=True)
plt.title("Confusion Matrix", fontsize=16)
plt.ylabel('Truth')
plt.xlabel('Prediction')
plt.show()
```

```python
from sklearn import metrics
```

```python
metrics.accuracy_score(y_test, predictions)
```

```python
metrics.precision_score(y_test, predictions)
```

```python
metrics.recall_score(y_test, predictions)
```
